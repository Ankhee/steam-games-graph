```python
import pandas as pd
import numpy as np
import scipy.spatial
df = pd.read_csv("C:\\Users\\Admin\\Documents\\GitHub\\GamesGraph\\scripts\\wip_data\\_dataframe_merged.csv",index_col=0)
df.head()
```

```python
df.fillna(0,inplace=True)
df.head()
```

```python
cosine_similarity = pd.DataFrame()
for index1, col1 in df.iteritems():
    row = {}
    u = col1.as_matrix()
    for index2, col2 in df.iteritems():
        v = col2.as_matrix()
        similarity = 1 - scipy.spatial.distance.cosine(u,v)
        row[index2] = similarity
    row_series = pd.Series(row, index=[index1])
    cosine_similarity = pd.concat([cosine_similarity, row_series])
    print(index1)
```

```python
cosine_similarity.to_csv("C:\\Users\\Admin\\Documents\\GitHub\\GamesGraph\\scripts\\wip_data\\_dataframe_merged_cosine_similarity.csv")
```
