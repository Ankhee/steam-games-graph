```python
import pandas as pd
import numpy as np
df = pd.read_csv("C:\\Users\\Admin\\Documents\\GitHub\\GamesGraph\\scripts\\wip_data\\_dataframe_merged.csv",index_col=0,dtype=np.float16)
df.head()
```

```python
df.drop('profile', inplace=True)
df.drop('time',inplace=True)
```

```python
game_users_treshold = 100
for index, col in df.iteritems():
    if col.count() < game_users_treshold:
        print(index,col.count(),'dropped!')
        df.drop(index, inplace=True)
```
