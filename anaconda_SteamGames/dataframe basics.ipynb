```python
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy
```

```python
#df = pd.read_csv("D:\\_WORK\\_MOJE\\201703 steam scraper\\steamscrapper\\scripts\\dataframe_final_ints.csv", index_col=0)
df_meta = pd.read_csv("dataframe_meta.csv", index_col=0, dtype={'url':str, 'total_hours_played':int}, encoding='utf-8')
df_values = pd.read_csv("dataframe_values_ints.csv", index_col=0, encoding='utf-8')
```

```python
df_values.fillna(0, inplace=True)
df_values = df_values.astype(int)
```

```python
df_values['Battlerite'].unique()
df_test = df_values['Battlerite']
print(df_test[0])
df_test.to_csv("dataframe_test.csv", encoding='utf-8')
```

```python
df = pd.concat([df_meta, df_values], axis=1)
df['Battlerite'][0]
```

```python
df.to_csv("dataframe_wip.csv", encoding='utf-8')
```

```python
#drop players with less than 10hs played

df = df[df.total_hours_played > 10]
print(df.shape)
```

```python
for index, col in df.iteritems():
    if type(col[0]) != numpy.int32:
        print(type(col[0]))
        continue
    if col.sum() < 10:
        print(index)
        df.drop(index, axis=1, inplace=True)
print(df.shape)
df.head()
```

```python
df.to_csv("dataframe_cleaned_170408.csv", encoding='utf-8')
```
